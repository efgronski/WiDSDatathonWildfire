## 📊 Distributions

Now this is where it easier for our brains to conceptualize patterns. We've been looking at a lot of tables, but that's boring as HELL! Let's make some graphs.

We are going to make some histograms. These show the distribution of data. We want to look for normal and non-normal distributions. It's time to bring in the trusty Bell Curve!

I'll make some, and then I want you to pick a new column and make some of your own.

```python
df["time_to_hit_hours"].hist()
plt.title("Time Until Fire Comes Within Evac Zone Distribution")
plt.show()
```

Well that's not a Bell Curve at all. What the heck is going on here? We see our time peak at the beginning, meaning lots of fires are coming within the evac zone fast. And then we see those numbers taper off, until the end. Why is that?

Well, as mentioned in the metadata:
- For censored events (never hit within 72h), this is the last observed time within the window (≤ 72).

So we have lots of events where we are unsure if the fire hits the evac zone because we stopped looking. So they just put the max time, 72 hours. Now this could be troublesome for our analysis, because we don't know if those fires ever hit the evac zone. Maybe some of them did after 72 hours, but maybe some of them didn't.

## 🌍 Relationships Between Variables
We want to compare groups in our data and see how they interact. This can help us discover relationships in the data. We will start doing this by using scatterplots and boxplots.

```python
sns.scatterplot(x="centroid_speed_m_per_h", y="area_growth_abs_0_5h", data=df)
```

```python
sns.scatterplot(x="centroid_speed_m_per_h", y="log1p_growth", data=df)
```

Now in these plots, I'm sure you can see one scatterplot is easier to read than the other. That's because one of them uses a log function!

Log transformations are applied to data that is not normally distributed to make it more symmetrical, helping data "behave better" for statistical tests. By reducing data spanning multiple orders of magnitude, the log function compresses large values and spreads out smaller ones, making patterns easier to identify.

```python
sns.regplot(x="centroid_speed_m_per_h", y="log1p_growth", ci=None, data=df)
```

And we can use the `regplot` function to make a line of best fit! As we can see, many of our points do not fall on our line, queueing us in that this relationship is probably not linear. Well, we don't have to have a straight line of best fit! We can change it up with the `order` parameter.

```python
sns.regplot(x="centroid_speed_m_per_h", y="log1p_growth", ci=None, order=2, data=df)
```

Oh! We can already see how much better this is. The distance between the points and the line seem smaller. However, we don't need to rely on our eyes to see if they are, we can do a quick calculation to measure this.

```python
sns.residplot(x="centroid_speed_m_per_h", y="log1p_growth", order=1, data=df)
sns.residplot(x="centroid_speed_m_per_h", y="log1p_growth", order=2, data=df)
```

Now we can see the residuals of our plots against each other! You can do this with any seaborn plots. In blue we have our linear relationship, and in orange we have our quadratic relationship. Looking at it this way, it is hard to tell which relationship is more representative. To find out we would need to calculate the actual values of these points. I won't tackle that here, but feel free to research and attempt it here in the notebook!
